````python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np

import pathlib

# folder to save figures
savefolder=pathlib.Path('./figures/')
````

## create some data

````python
x = np.linspace(0, 2, 100)
y1 = x**2
y2 = x**3


rng = np.random.default_rng()
z = rng.normal(0,1,100)
````

# How to improve matplotlib

## Default Matplotlib

````python
plt.figure()
plt.plot(x,x,label='linear')
plt.plot(x,y1, label='quadratic')
plt.plot(x,y2, label='cubic')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
````

````python
# plt.style.use('default')
fig, ax = plt.subplots()
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, y1, label='quadratic')  # Plot more data on the axes...
ax.plot(x, y2, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()

# comment two lines below to stop figure saving
f = savefolder / 'example1.png'
fig.savefig(f, bbox_inches='tight')

plt.show() 
````

````python
from scipy import stats

# estimate distribution using kde
# fit kde to z
kde = stats.gaussian_kde(z)
# visualise across distribution
v = np.linspace(z.min(), z.max(), 100)
est = kde(v)

fig,ax =plt.subplots()
ax.hist(z, bins=10, density=True, alpha=0.5, label='z')
ax.plot(v, est, color='C0', linestyle='-', label='kde')
ax.legend()
ax.set_title('Simple Histogram')
f = savefolder / 'example2.png'
fig.savefig(f, bbox_inches='tight')
plt.show()
````

````python
rng = np.random.default_rng(seed=2)
x1 = rng.normal(0,1,100)
x2 = rng.normal(0,1,100)
x3 = rng.normal(0,1,100)

fig, ax = plt.subplots(figsize=(4,2))
ax.plot(np.arange(len(x1)), x1, label='x1')
ax.plot(np.arange(len(x2)), x2, label='x2')
ax.plot(np.arange(len(x3)), x3, 'd', label='x3')
ax.legend(loc='upper right');
f = savefolder / 'example3_small.png'
fig.savefig(f, bbox_inches='tight',dpi=50)
plt.show()
````

## DPI vs Figsize

Example taken from [stackoverflow](https://stackoverflow.com/questions/47633546/relationship-between-dpi-and-figure-size)

````python
import matplotlib.pyplot as plt
%matplotlib inline

def plot(fs,dpi, savepath):
    fig, ax=plt.subplots(figsize=fs, dpi=dpi)
    ax.set_title("Figsize: {}, dpi: {}".format(fs,dpi))
    ax.plot([2,4,1,5], label="Label")
    ax.legend()
    # fig.tight_layout()
    fig.savefig(savepath)
    fig.show()

figsize=(2,2)
for i in range(1,4):
    f = savefolder / f'dpi_{i*72}.png'
    plot(figsize, i*72, f)

dpi=72
for i in [2,4,6]:
    f = savefolder / f'figsize_{i}_{i}.png'
    plot((i,i), dpi, f)
````

## Reduce clutter

````python
rng = np.random.default_rng()

x = np.linspace(0, 2, 100)
ys = []
for i in range(5):
    y = rng.normal(0,1,100)
    ys.append(y)
````

````python
plt.style.use('default')

fig, ax = plt.subplots(dpi=150)
for i, y in enumerate(ys):
    ax.plot(x,y,label=f'Series {i+1}')
ax.legend()
f = savefolder / 'toomuchclutter.png'
fig.savefig(f, dpi=300)
plt.show()

# better
fig, axs = plt.subplots(nrows=len(ys), dpi=150, sharex=True)
for i in range(len(ys)):
    ax = axs[i]
    for j,y in enumerate(ys):
        if i==j:
            ax.plot(x,y, color='C0', label=f'Series {j}')
        else:
            ax.plot(x,y, color='C0', alpha=0.2)
    
    ax.set_ylabel(f'Series {i}', rotation=0, labelpad=20)
    ax.set_yticks([])
    ax.grid(True, axis='x', color='#9E9E9E')
    ax.spines['top'].set_visible(False)
    # ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)


fig.tight_layout()
f = savefolder / 'reduce_clutter.png'
fig.savefig(f, dpi=300)
plt.show()
````

## XKCD plot

````python
import numpy as np

import matplotlib
# matplotlib.use('Agg')
import matplotlib.pyplot as plt
plt.xkcd();

X = np.linspace(0,2.32,100)
Y = X*X - 5*np.exp(-5*(X-2)*(X-2))

fig = plt.figure(figsize=(12,5), dpi=72,facecolor="white")
axes = plt.subplot(111)

plt.plot(X,Y, color = 'k', linewidth=2, linestyle="-", zorder=+10)

axes.set_xlim(X.min(),X.max())
axes.set_ylim(1.01*Y.min(), 1.01*Y.max())

axes.spines['right'].set_color('none')
axes.spines['top'].set_color('none')
axes.xaxis.set_ticks_position('bottom')
axes.spines['bottom'].set_position(('data',0))
axes.yaxis.set_ticks_position('left')
axes.spines['left'].set_position(('data',X.min()))

axes.set_xticks([])
axes.set_yticks([])
axes.set_xlim( 1.05*X.min(), 1.10*X.max() )
axes.set_ylim( 1.15*Y.min(), 1.05*Y.max() )

t = [10,40,82,88,93,99]
plt.scatter( X[t], Y[t], s=50, zorder=+12, c='k')

plt.text(X[t[0]]-.1, Y[t[0]]+.1, "Industrial\nRobot", ha='left', va='bottom')
plt.text(X[t[1]]-.15, Y[t[1]]+.1, "Humanoid\nRobot", ha='left', va='bottom')
plt.text(X[t[2]]-.25, Y[t[2]], "Zombie", ha='left', va='center')
plt.text(X[t[3]]+.05, Y[t[3]], "Prosthetic\nHand", ha='left', va='center')
plt.text(X[t[4]]+.05, Y[t[4]], "Bunraku\nPuppet", ha='left', va='center')
plt.text(X[t[5]]+.05, Y[t[5]], "Human", ha='left', va='center')
plt.text(X[t[2]]-0.05, 1.5, "Uncanny\nValley", ha='center', va='center', fontsize=24)

plt.ylabel("-      Comfort Level      +",y=.5, fontsize=20)
plt.text(.05, -.1, "Human Likeness ->",ha='left', va='top', color='r', fontsize=20)

X = np.linspace(0,1.1*2.32,100)
axes.fill_between(X, 0, -10, color = '0.85', zorder=-1)
axes.fill_between(X, 0, +10, color = (1.0,1.0,0.9), zorder=-1)

#X = np.linspace(1.652,2.135,100)
X = np.linspace(1.5,2.25,100)
Y = X*X - 5*np.exp(-5*(X-2)*(X-2))
axes.fill_between(X, Y, +10, color = (1,1,1), zorder=-1)

axes.axvline(x=1.5,ymin=0,ymax=1, color='.5', ls='--')
axes.axvline(x=2.25,ymin=0,ymax=1, color='.5', ls='--')

# plt.savefig("figure-8.pdf")
plt.show()
````
